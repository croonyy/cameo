```python
from IPython.display import HTML, display
def generate_arcs(start_r, n, start_point):
    # 🔥 4个方向循环：[x加减, y加减] +1=加  -1=减
    directions = [
        (1,  -1),   # 0: + -
        (-1, -1),   # 1: - -
        (-1, 1),    # 2: - +
        (1,  1)     # 3: + +
    ]
    
    x, y = start_point  # 当前坐标
    r_prev2, r_prev1 = start_r  # 前两个半径
    lines = []
    
    for i in range(n):
        # 1. 取当前半径（斐波那契）
        if i == 0:
            r = r_prev2
        elif i == 1:
            r = r_prev1
        else:
            r = r_prev2 + r_prev1
            r_prev2, r_prev1 = r_prev1, r  # 更新前两个半径
        
        # 🔥 2. 循环取方向（核心！）
        dx, dy = directions[i % 4]
        
        # 3. 计算下一个坐标（按方向 +- 半径长度）
        next_x = x + dx * r
        next_y = y + dy * r
        
        # 4. 生成指令行
        line = f"        M{x} {y} A{r} {r} 0 0 1 {next_x} {next_y}\n        M{x} {y} A{r} {r} 0 0 0 {next_x} {next_y}"
        if i > 0:
            lines.append(line)
        
        # 5. 更新起点为下一个点
        x, y = next_x, next_y
    
    return "\n".join(lines)

# ===================== 调用 =====================
w = 340  # 正方形画布大小
r = [21, 34]  # 前两个半径
n = 5 # 几段弧
s = [121, 160] # 开始点
# c = ["#000000", "#00ffff", "#ffffff"]
# c = ["#D81B60", "#80DEEA", "#ffffff"]
c = ["#00c1d4", "#000000", "#ffffff"]

result = generate_arcs(r, n, s)
html= f"""\
<svg width="{w}" height="{w}" viewBox="0 0 {w} {w}" xmlns="http://www.w3.org/2000/svg">
    <rect width="{w}" height="{w}" fill="#ffffff" />
    <path
        d="
{result}
        "
        fill="none" stroke="{c[0]}"
        stroke-width="100"
        stroke-linecap="round" />
    <path
        d="
{result}
        "
        fill="{c[1]}" stroke="{c[0]}"
        stroke-width="21"
        stroke-linecap="round" />
    <circle cx="{s[0]}" cy="{s[1]}" r="25" fill="{c[2]}" />
</svg>
"""
print(html)
display(HTML(html))
```
