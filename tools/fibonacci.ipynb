```python
from IPython.display import HTML, display

w = 600 # 画布宽高
unit = 37 # 基础单位，斐波那契数列的第 20 项为 6765，除以这个数得到一个合适的缩放比例
c = ["#00ffff", "#012772", "#00ffff"] # 颜色：线条颜色、叶子颜色、点的颜色

# 斐波那契数列切片函数，返回从第 start 项开始的 count 项
def fib_slice(start, count):
    res = [0, 1]
    max_len = start + count - 1
    for i in range(2, max_len):
        res.append(res[i - 1] + res[i - 2])
    return res[start - 1 : start - 1 + count]


# 从第 4 项开始，取 6 项
fi = fib_slice(20, 5)
fi_mod = [round(x / fi[0] * unit, 3) for x in fi]


# 4个方向循环：[x加减, y加减] +1=加  -1=减
directions = [
    (1, -1),  # 0: + -
    (-1, -1),  # 1: - -
    (-1, 1),  # 2: - +
    (1, 1),  # 3: + +
]

# 计算图形最大最小x,y 计算中心点偏移量
x,y=0,0
min_x,max_x,min_y,max_y = 0,0,0,0
for i, r in enumerate(fi_mod):
    dx, dy = directions[i % 4]
    next_x = round(x + dx * r, 3)
    next_y = round(y + dy * r, 3)
    x, y = next_x, next_y
    min_x,max_x,min_y,max_y = min(min_x, next_x), max(max_x, next_x), min(min_y, next_y), max(max_y, next_y)

# sp = (w/2, w/2) # 画布中心点
sp = (w/2 - (min_x+max_x)/2, w/2 - (min_y+max_y)/2) # 画布中心补偿偏移量的图形起始点坐标
x, y = sp
leafs = ""
dots = ""
for i, r in enumerate(fi_mod):
    dx, dy = directions[i % 4]
    # 3. 计算下一个坐标（按方向 +- 半径长度）
    next_x = round(x + dx * r, 3)
    next_y = round(y + dy * r, 3)
    leaf = (round(x+dx*r/2, 3), round(y+dy*r/2, 3))
    # print(x,y, next_x, next_y)
    # 5. 更新起点为下一个点
    if i > 0:
        leafs += f"M{x} {y} A{r} {r} 0 0 1 {next_x} {next_y}\nM{x} {y} A{r} {r} 0 0 0 {next_x} {next_y}\n" 
        dots += f'<circle cx="{leaf[0]}" cy="{leaf[1]}" r="{round(r/10, 3)}" />\n'
    x, y = next_x, next_y

html = f"""<svg width="{w}" height="{w}" viewBox="0 0 {w} {w}" xmlns="http://www.w3.org/2000/svg">
    <rect width="{w}" height="{w}" fill="#ffffff" />
    <path
        d="{leafs}"
        fill="none" stroke="{c[0]}"
        stroke-width="{unit*5}"
        stroke-linecap="round" />
    <path
        d="{leafs}"
        fill="{c[1]}" stroke="{c[0]}"
        stroke-width="0"
        stroke-linecap="round" />
    <circle cx="{sp[0]}" cy="{sp[1]}" r="20" fill="#012772">
        <animate attributeName="r" values="{unit/2};{unit};{unit/2}" dur="1.5s" repeatCount="indefinite" />
    </circle>
    <g fill="{c[2]}">{dots}</g>
</svg>
"""
print(html)
display(HTML(html))
```
